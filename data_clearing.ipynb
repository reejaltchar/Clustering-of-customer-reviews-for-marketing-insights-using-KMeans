```python
#importing the packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
#import dataset
dataset = pd.read_csv('amazon_review.csv')
dataset
```

```python
#data preorocessing
print(dataset.columns)
```

```python
# Overview of missing values
print(dataset.isna().sum())
```

```python
# Drop columns not critical to analysis
dataset = dataset.drop(columns=['dimension', 'ean', 'manufacturerNumber', 'upc', 'weight'], errors='ignore')
dataset
```

```python
# Fill categorical columns
dataset['colors'] = dataset['colors'].fillna('Unknown')
dataset['reviews.doRecommend'] = dataset['reviews.doRecommend'].fillna('Unknown')
dataset['reviews.title'] = dataset['reviews.title'].fillna('No Title')
dataset['reviews.username'] = dataset['reviews.username'].fillna('Anonymous')

# Fill numerical columns
dataset['reviews.numHelpful'] = dataset['reviews.numHelpful'].fillna(0)
dataset['reviews.rating'] = dataset['reviews.rating'].fillna(dataset['reviews.rating'].mean())
```

```python
dataset= dataset.dropna(subset=['reviews.date', 'manufacturer'])
dataset.info()
```
